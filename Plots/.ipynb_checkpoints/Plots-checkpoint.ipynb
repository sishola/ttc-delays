```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import datetime
import os

# the program 'HeroesOfPymoli.ipynb' file path will need to be determined
path = os.path.dirname(os.path.abspath('Plots.ipynb'))
#print(path)

# set the path for purchase_data.csv file to be read
subway_csv = os.path.join(path, 'subway_master.csv')
surface_csv = os.path.join(path, 'surface_master.csv')
surface_station_csv = os.path.join(path, 'ttc_Delays_surfaceRoutes_cleaned.csv')

# Read Purchasing File and store into Pandas data frame
subway = pd.read_csv(subway_csv)
surface = pd.read_csv(surface_csv)
surface_station = pd.read_csv(surface_station_csv, encoding='cp1252')
```

```python
surface_station.head()
```

```python
streetcar = surface_station.loc[(surface_station["Type"] == "Streetcar") & (surface_station["Min Delay"] < 40) \
                               & (surface_station["Min Delay"] > 0)]
streetcar.reset_index(drop=True, inplace=True)
streetcar.head(40)
```

```python
streetcar.dtypes
```

Morning Peak Time : 6:30 am - 10 am
Day : 10 am - 3:30 pm
Afternoon Peak Time : 3:30 pm - 7 pm
Off Hours : 7 pm - 6:30 am
https://www.blogto.com/city/2014/04/when_exactly_is_rush_hour_in_toronto/

```python
bins = [0, 6.5, 10, 15.5, 19, 24]
group_names = ["Off Hours1", "Morning Peak Time", "Day Time", "Afternoon Peak Time", "Off Hours2"]
#hours_week_labels = 24
#hours_week = 25
#for i in range(hours_week):
#    bins.append(i)
    
#for j in range(hours_week_labels):
#    j += 1
#    group_names.append(j)

streetcar["Hour Summary"] = pd.cut(streetcar["Time"], bins, labels=group_names)

conditions = [
    (streetcar["Day"] == "Monday") | (streetcar["Day"] == "Tuesday") | (streetcar["Day"] == "Wednesday") | \
    (streetcar["Day"] == "Thursday") | (streetcar["Day"] == "Friday"), 
    (streetcar["Day"] == "Saturday") | (streetcar["Day"] == "Sunday")]

choices = ["Weekday", "Weekend"]

streetcar["Weekday"] = np.select(conditions, choices, default=None)

streetcar["Hour Summary"].replace("Off Hours1", "Off Hours", inplace=True)
streetcar["Hour Summary"].replace("Off Hours2", "Off Hours", inplace=True)

streetcar.head(100)
```

```python
streetcar.Weekday.unique()
```

```python
streetcar_route = streetcar.groupby(['Weekday', "Hour Summary"]).agg({'Min Delay': ['count', 'mean', "median", 'sum', 'sem'], 'Min Gap': ['mean', 'sum', 'sem']})
streetcar_route_low = streetcar.groupby(['Weekday', "Hour Summary"]).quantile(q=0.25)
streetcar_route_high = streetcar.groupby(['Weekday', "Hour Summary"]).quantile(q=0.75)
streetcar_route_low.drop(columns=["Unnamed: 0", "Route", "Time", "Min Gap", "Vehicle", "Year"], inplace=True)
streetcar_route_low = streetcar_route_low.reset_index(level=['Weekday', 'Hour Summary'])
```

```python
weekday = streetcar_route_low["Weekday"].tolist()
hour = streetcar_route_low["Hour Summary"].tolist()
quart_1 = streetcar_route_low["Min Delay"].tolist()
quart_3 = streetcar_route_high["Min Delay"].tolist()
median = streetcar_route["Min Delay"]["median"].tolist()
mean = streetcar_route["Min Delay"]["mean"].tolist()
error = streetcar_route["Min Delay"]["sem"].tolist()


dict1 = {"Weekday": weekday, 
        "Time of Day": hour, 
        "Mean": mean, 
        "Standard Error": error, 
        "Median": median, 
        "0.25 Quartile": quart_1, 
        "0.75 Quartile": quart_3}

combined = pd.DataFrame(dict1)
combined
```

```python
combined2 = combined.reindex([2, 1, 0, 3, 6, 5, 4, 7])
combined2.reset_index(inplace=True)
combined2.drop(columns="index", inplace=True)
combined2
```

```python
streetcar.boxplot("Min Delay", by=["Weekday", "Hour Summary"], figsize=(20, 10), showfliers=False, showmeans=True, autorange=True)
```
